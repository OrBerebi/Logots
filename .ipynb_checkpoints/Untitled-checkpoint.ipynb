```python
import pandas as pd
import ast
import numpy as np
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('data.csv')

# Convert the string representations of lists into actual Python lists
df['yaw'] = df['yaw'].apply(ast.literal_eval)
df['pitch'] = df['pitch'].apply(ast.literal_eval)
df['roll'] = df['roll'].apply(ast.literal_eval)

# Plot Unwrapped Yaw Values
plt.figure(figsize=(8, 4))
for idx, row in df.iterrows():
    # Convert the yaw list to a NumPy array
    yaw = np.array(row['yaw'])
    # Convert from degrees to radians, unwrap the angle, then convert back to degrees
    yaw_unwrapped = np.rad2deg(np.unwrap(np.deg2rad(yaw)))
    plt.plot(yaw_unwrapped, marker='o', label=f"Frame {row['frame_id']}")

plt.xlabel("Sample Index")
plt.ylabel("Yaw (degrees)")
plt.title("Unwrapped Yaw")
plt.legend()
plt.tight_layout()
plt.show()

# Plot Pitch Values
plt.figure(figsize=(8, 4))
for idx, row in df.iterrows():
    pitch = np.array(row['pitch'])
    plt.plot(pitch, marker='o', label=f"Frame {row['frame_id']}")
    
plt.xlabel("Sample Index")
plt.ylabel("Pitch (degrees)")
plt.title("Pitch")
plt.legend()
plt.tight_layout()
plt.show()

# Plot Roll Values
plt.figure(figsize=(8, 4))
for idx, row in df.iterrows():
    roll = np.array(row['roll'])
    plt.plot(roll, marker='o', label=f"Frame {row['frame_id']}")
    
plt.xlabel("Sample Index")
plt.ylabel("Roll (degrees)")
plt.title("Roll")
plt.legend()
plt.tight_layout()
plt.show()
```
